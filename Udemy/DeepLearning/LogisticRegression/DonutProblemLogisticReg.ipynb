```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return (1/(1+ np.exp(-z)))

def forward(X, W):
    return sigmoid(X.dot(W))

def cross_entropy_cost(T, Y):
    return -(T * np.log(Y) + (1-T)* np.log(1-Y)).mean()

def classification_rate(T, Y):
    return 1 - np.abs(T - np.round(Y)).sum() / N
```

```python
N = 1000
D = 2

R_inner = 5
R_outer = 10

R1 = np.random.randn(N /2) + R_inner
theta1 = 2*np.pi*np.random.random(N/2)
X_inner = np.concatenate([[R1 * np.cos(theta1)], [R1 * np.sin(theta1)]]).T

R2 = np.random.randn(N /2) + R_outer
theta2 = 2*np.pi*np.random.random(N/2)
X_outer = np.concatenate([[R2 * np.cos(theta2)], [R2 * np.sin(theta2)]]).T

X = np.concatenate([X_inner, X_outer])
T = np.array([0] * (N/2) + [1] * (N/2))


plt.scatter(X[:, 0], X[:, 1], c = T)
#plt.show()

bias = np.array([[1] * N]).T
# Trick to solve donut problem is to calculate radius of all the points

radius = np.zeros((N,1))

# Calculate radius of all the points manually

for i in xrange(N):
    radius[i] = np.sqrt(X[i,:].dot(X[i,:]))
    
print radius.shape
print X.shape
print bias.shape

Xb = np.concatenate((bias, radius, X), axis=1)
print Xb.shape
```

```python
W = np.random.randn(D + 2)
learning_rate = 0.001
l2_reg = 0.01
errors = []
for i in xrange(5000):
    Y = forward(Xb, W)
    cost = cross_entropy_cost(T, Y)
    errors.append(cost)
    #print W.shape
    #print T.shape
    #print Y.shape
    W += learning_rate * (Xb.T.dot(T - Y) - l2_reg * W)
    #print W
    print "Classification rate: ", classification_rate(T, Y)
    

plt.plot(errors)
plt.title("Cross Entropy errors")
plt.show()
```
