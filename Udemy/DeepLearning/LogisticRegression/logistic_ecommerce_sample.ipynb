```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
def get_data():
    df = pd.read_csv('ecommerce_data.csv')
    data = df.as_matrix()
    print df.head()
    X = data[:, : -1]
    Y = data[:, -1] 
    
    X[:, 1] = (X[:,1] - X[:,1].mean()) / X[:,1].std()
    X[:, 2] = (X[:,2] - X[:,2].mean()) / X[:,2].std()
    
    N, D = X.shape
    X2 = np.zeros((N, D+3))
    X2[:, : D-1] = X[:, : D-1]
    
    for n in xrange(N):
        t = X[n, D-1]
        X2[n, t+ D-1] = 1
    
    #print X2
    
    Z = np.zeros((N, 4))
    Z[np.arange(N), X[:, D-1].astype(np.int32)] = 1
    #print Z
    
    assert(np.abs(X2[:, -4:] - Z).sum() < 10e-10)
    return X2, Y

def get_binary_data():
    X, Y = get_data()
    X2 = X[Y <= 1]
    Y2 = Y[Y <= 1]
    return X2, Y2
```

```python
X, Y = get_binary_data()
print X.shape
X, Y = shuffle(X, Y)
X_train = X[:-100]
X_test = X[-100:]
T_train = Y[:-100]
T_test = Y[-100:]

print X_train.shape
D = X.shape[1]
W = np.random.randn(D)
b = 0

def sigmoid(z):
    return (1/(1+ np.exp(-z)))

def forward(X, W, b):
    return sigmoid((X.dot(W) + b))

def classifier_rate(T, Y):
    return np.sum(T == Y)

def cross_entropy_cost(T, Y):
    return -np.sum((T * np.log(Y) + (1-T)* np.log(1-Y)))

learning_rate = 0.001

train_costs = []
test_costs = []
for i in xrange(1000):
        Y_train = forward(X_train, W, b)
        Y_test = forward(X_test, W, b)
        train_cost = cross_entropy_cost(T_train, Y_train)
        test_cost = cross_entropy_cost(T_test, Y_test)
        train_costs.append(train_cost)
        test_costs.append(test_cost)
        W -= learning_rate * X_train.T.dot(Y_train - T_train)
        b -= learning_rate * ((Y_train - T_train).sum())
        if (i % 1000):
            print "id %i Train cost : %d and Test Cost: %d" %(i, train_cost, test_cost)
            
rate_train = classifier_rate(T_train, np.round(Y_train))            
rate_test = classifier_rate(T_test, np.round(Y_test))
print "Classifier rate test %d" %rate_test
print "Classifier rate train %d" %rate_train
```
